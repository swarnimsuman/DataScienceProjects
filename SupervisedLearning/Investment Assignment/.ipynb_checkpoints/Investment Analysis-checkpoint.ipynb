## Checkpoint 4

```python
top9 = top9.reset_index()
```

```python
top9['category_list'] = top9['category_list'].str.split('|', 2).str[0]
top9['category_list'].value_counts()
```

```python
mappings = mappings.set_index('category_list')
mappings
```

```python
mappings.columns.tolist()
```

```python
updated_mappings = pd.DataFrame() 
for sector in mappings.columns.tolist():
    sector_mapping = pd.DataFrame(mappings.index[mappings[sector] == 1].tolist(), columns=['category_list'])
    sector_mapping['sector'] = sector
    updated_mappings = updated_mappings.append(sector_mapping)

updated_mappings.reset_index()
updated_mappings
```

```python
updated_mappings = updated_mappings.apply(lambda x: x.astype(str).str.lower())
updated_mappings
```

```python
# Merge Both the dataframes
top9 = top9.merge(updated_mappings, on='category_list', how='inner')
top9.head()
```

```python
del top9['index']
```

```python
top9.head()
```

## Checkpoint 6

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.style.use("ggplot")
```

#### 1. The representative amount of investment in each funding type

```python
master_frame['raised_amount_in_millions'] = master_frame['raised_amount_usd'].apply(lambda x: x/1000000)
```

```python
sns.set(rc={'figure.figsize':(15.7,12.27)})
plt.ylim(0, 15000000)
bar_plt = sns.barplot(data=master_frame, x='funding_round_type', y='raised_amount_usd', estimator=np.mean, order=['venture', 'seed', 'private_equity', 'angel'])
bar_plt.set(xlabel = "Funding Type", ylabel = "Amount Raised In Millions(USD)")
plt.xticks(rotation=70)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15.7,12.27)})
plt.ylim(0, 50)
bar_plt = sns.barplot(data=master_frame, x='funding_round_type', y='raised_amount_in_millions', estimator=np.mean, palette="Greens", order=['venture', 'seed', 'private_equity', 'angel'])
bar_plt.set(xlabel = "Funding Type", ylabel = "Amount Raised In Millions(USD)")
plt.xticks(rotation=70)
plt.show()
```
